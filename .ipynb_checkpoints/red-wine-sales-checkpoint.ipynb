# Codebar - Predicting the Future in Python

An example notebook for my lightening talk on time series forecasting in python, built for the Deliveroo Codebar event 27th April 2016.

## 1. Munge the Data

```python
# Import stuff we need
import os
import math
import pandas as pd
import plotly.plotly as py
import plotly.graph_objs as go
from plotly import tools
import pandas as pd
import statsmodels.api as sm
import datetime as dt
from sklearn.linear_model import LinearRegression
import numpy as np
from dateutil.relativedelta import relativedelta
```

```python
# Check what the data file is called
os.listdir(os.getcwd())
```

```python
# Read it into pandas and check it looks ok
data = pd.read_csv('red-wine.csv')
data.head()
```

```python
# Parse the dates into something pandas can understand
data.columns = ['period', 'sales_volume']
data.loc[:, 'biz_date'] = data.period.apply(lambda x: dt.datetime.strptime(x, '%b-%y').date())
data_clean = data.loc[:, ['biz_date','sales_volume']]
data_clean.head()
```

```python
# Create a plot - lets look at it
trace = go.Scatter( x=data_clean['biz_date'], y=data_clean['sales_volume'] )
plot_data = [trace]

layout = go.Layout(
    yaxis=dict(
        title='Units Sold'
    ),
    title='Australian Red Wine Sales to USA'
)

figure = go.Figure(data=plot_data, layout=layout)

# py.iplot(
#     figure, 
#     filename='codebar/red-wine-sales'
# )

py.image.ishow(figure)
```

## 2. Decompose the Time Series

```python
# Perform the decomposition <<< MAGIC HAPPENS HERE
res = sm.tsa.seasonal_decompose(data_clean.sales_volume.values, freq=12)
```

```python
# Extract the three components
trend = res.trend
seasonal = res.seasonal
resid = res.resid
```

```python
# Make into dataframes so we can join
trend_frame = pd.DataFrame(trend, index=data_clean.biz_date, columns=['trend']).fillna(0)
seasonal_frame = pd.DataFrame(seasonal, index=data_clean.biz_date, columns=['seasonal']).fillna(0)
resid_frame = pd.DataFrame(resid, index=data_clean.biz_date, columns=['resid']).fillna(0)
```

```python
# Merge them all back into one frame for eas of use
data_original = data_clean.copy()
data_original = data_original.set_index('biz_date')
data_merged = pd.merge(data_original, trend_frame, left_index=True, right_index=True)
data_merged = pd.merge(data_merged, seasonal_frame, left_index=True, right_index=True)
data_merged = pd.merge(data_merged, resid_frame, left_index=True, right_index=True)
data_merged.head(10)
```

```python
# Plot again to see what we got!
trace1 = go.Scatter(x=data_merged.index, y=data_merged['sales_volume'], name='Sales Volume')
trace2 = go.Scatter(x=data_merged.index, y=data_merged['trend'], name='Trend')
trace3 = go.Scatter(x=data_merged.index, y=data_merged['seasonal'], name='Seasonal')
trace4 = go.Scatter(x=data_merged.index, y=data_merged['resid'], name='Residual')

fig = tools.make_subplots(rows=4, cols=1);

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 2, 1)
fig.append_trace(trace3, 3, 1)
fig.append_trace(trace4, 4, 1)


fig['layout'].update(height=800, width=1000, title='Decomposed Timeseries')

# py.iplot(
#     fig, 
#     filename='codebar/decomposed-raw'
# )

py.image.ishow(fig)
```

```python
# Remove the lead / trail zeros, add in mid steps and plot one more time
decomposed_clean = data_merged.copy()
decomposed_clean = decomposed_clean.loc[decomposed_clean.trend != 0]
decomposed_clean.loc[:,'sales_minus_trend'] = decomposed_clean.sales_volume - decomposed_clean.trend
decomposed_clean.loc[:,'trend_minus_seasonal'] = decomposed_clean.sales_volume - decomposed_clean.trend - decomposed_clean.seasonal


trace1 = go.Scatter(x=decomposed_clean.index, y=decomposed_clean['sales_volume'], name='Sales Volume')
trace2 = go.Scatter(x=decomposed_clean.index, y=decomposed_clean['trend'], name='Trend')
trace3 = go.Scatter(x=decomposed_clean.index, y=decomposed_clean['sales_minus_trend'], name='Sales Minus Trend')
trace4 = go.Scatter(x=decomposed_clean.index, y=decomposed_clean['seasonal'], name='Seasonal')
trace5 = go.Scatter(x=decomposed_clean.index, y=decomposed_clean['trend_minus_seasonal'], name='Sales Minus Seasonal Minus Trend')
trace6 = go.Scatter(x=decomposed_clean.index, y=decomposed_clean['resid'], name='Residual')

fig = tools.make_subplots(rows=6, cols=1);

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 2, 1)
fig.append_trace(trace3, 3, 1)
fig.append_trace(trace4, 4, 1)
fig.append_trace(trace5, 5, 1)
fig.append_trace(trace6, 6, 1)


fig['layout'].update(height=1100, width=1000, title='Decomposed Timeseries')

# py.iplot(
#     fig, 
#     filename='codebar/decomposed-clean'
# )
py.image.ishow(fig)
```

## Forecast Each Component

```python
# First lets forecast the trend for the next 5 years
obs_numbers = decomposed_clean.reset_index().reset_index().loc[:, ['index']]
trend = decomposed_clean.trend.values


forecast_window = np.array([[x] for x in range(obs_numbers.index.max() + 1,obs_numbers.index.max() + 61)])
lm_lin = LinearRegression().fit(obs_numbers, trend)
forecast_trend = lm_lin.predict(forecast_window)

print(forecast_trend)
```

```python
# Now lets cycle the seasonal and forecast that 60 steps forewwards too

# THIS IS A FAFF!!!

all_seasonal = decomposed_clean.seasonal.values
ending_val = all_seasonal[-1:]
for i in range(0, len(all_seasonal)-1):
    if all_seasonal[i] == ending_val:
        break
base_seasonal_forecast_pattern = all_seasonal[i: i + 12]
extended_seasonal_forecast_pattern = np.tile(base_seasonal_forecast_pattern,
                                             math.ceil(60 / 7.0))
forecast_seasonal = extended_seasonal_forecast_pattern[0: 60]
print(forecast_seasonal)
```

```python
# Combine to get the forecast
forecast_vals = forecast_trend + forecast_seasonal
print(forecast_vals)
```

```python
# Again, lets mash these back into a dataframe
new_months = []
for i in range(1,61):
    new_months.append(decomposed_clean.index.max() + relativedelta(months=i))
    
combined_data = [list(forecast_trend), list(forecast_seasonal), list(forecast_vals)]
forecast = pd.DataFrame(combined_data).transpose()
forecast = forecast.set_index(pd.DatetimeIndex(new_months))
forecast.columns = ['trend_forecast', 'seasonal_forecast', 'forecast']
forecast.head()
```

```python
# Extend the original dataframe so we bring it all back together
combined_forecast = pd.concat([decomposed_clean, forecast])
combined_forecast.head()
```

```python
# Lets uplift the seasonal so we can plot it obviously ...
last_real_sales_val = combined_forecast.sales_volume.dropna().values[-1]
combined_forecast.loc[:, 'seasonal_forecast_uplift'] = combined_forecast['seasonal_forecast'] + last_real_sales_val
```

```python
# Plot the final result
forecast_plot_data = [
    go.Scatter(x=combined_forecast.index, y=combined_forecast['sales_volume'], name='Sales Volume'),
    go.Scatter(x=combined_forecast.index, y=combined_forecast['trend_forecast'], name='Trend Forecast'),
    go.Scatter(x=combined_forecast.index, y=combined_forecast['seasonal_forecast_uplift'], name='Seasonal Forecast'),
    go.Scatter(x=combined_forecast.index, y=combined_forecast['forecast'], name='Forecast')
]

layout = go.Layout(
    yaxis=dict(
        title='Units Sold'
    ),
    title='Forecast - Australian Red Wine Sales to USA',
    height=700, 
    width=1000
)

fig = go.Figure(data=forecast_plot_data, layout=layout)

# py.iplot(
#     fig, 
#     filename='codebar/forecast'
# )

py.image.ishow(fig)
```
